## 2. Prisoner's Dilemma — Payoff Matrix

Classic game theory example:

```python
# Prisoner's Dilemma Payoff Matrix
# Rows: Player A (Cooperate / Defect)
# Cols: Player B (Cooperate / Defect)
# Values: (A_payoff, B_payoff)

payoff_matrix = {
    ('Cooperate', 'Cooperate'): (3, 3),
    ('Cooperate', 'Defect'): (0, 5),
    ('Defect', 'Cooperate'): (5, 0),
    ('Defect', 'Defect'): (1, 1)
}

print("Prisoner's Dilemma Payoff Matrix:")
print("=" * 50)
print("                Player B")
print("          Cooperate    Defect")
print("Player A")
for a_action in ['Cooperate', 'Defect']:
    row = f"{a_action:10}"
    for b_action in ['Cooperate', 'Defect']:
        payoff = payoff_matrix[(a_action, b_action)]
        row += f"   {payoff}"
    print(row)

print("\nNash Equilibrium: (Defect, Defect) — both players betray")
print("Pareto Optimal: (Cooperate, Cooperate) — but not stable")
```

## 3. Case 1: Hong Kong Property Market (地產商)

Three major players: New World, CK Asset, Henderson

Scenario: All decide on pricing strategy simultaneously

```python
# HK Property Market Payoff Matrix (Profit in HK$b)
# High Price = Both keep high margins
# Low Price = Compete for market share

property_game = pd.DataFrame({
    'New World \ CK Asset': ['High Price (10,10)', 'Low Price (5,8)'],
    'High Price': ['(10, 10) Both win', '(8, 5) CK wins'],
    'Low Price': ['(5, 8) NW wins', '(3, 3) Both lose']
})

print("HK Property Market — Pricing Game:")
print("=" * 60)
print(property_game.to_string(index=False))
print("\nNash Equilibrium: (High Price, High Price)")
print("Result: High property prices sustained in HK")
```

## 4. Case 2: Hong Kong Banking (銀行)

HSBC, Hang Seng, Bank of China — Interest rate competition

```python
# Banking Interest Rate Game
# All banks have Dominant Strategy: Follow rate cuts

banking_payoff = pd.DataFrame({
    'HSBC \ Hang Seng': ['Keep Rate (8,8)', 'Cut Rate (5,10)'],
    'Keep Rate': ['(8, 8) Both stable', '(10, 5) HSBC loses'],
    'Cut Rate': ['(5, 10) HSB gains', '(3, 3) All lose profit']
})

print("HK Banking — Interest Rate Game:")
print("=" * 60)
print(banking_payoff.to_string(index=False))
print("\nDominant Strategy: Cut Rate (regardless of opponent)")
print("Result: HK banks engage in rate competition → lower margins")
```

## 5. Case 3: Telecom Industry (電訊商)

CMHK, 3HK, SmarTone — 5G pricing war

Repeated game: Short-term betrayal, long-term Tit-for-Tat

```python
# Telecom Repeated Game Analysis

# Round 1: All cooperate (High prices)
# Round 2: One betrays (low price to steal market)
# Round 3+: Tit-for-Tat strategy emerges

print("Telecom 5G Price War — Repeated Game:")
print("=" * 60)
print("Round 1: All charge HK$298/month for 5G")
print("Round 2: CMHK cuts to HK$198 — gains market share")
print("Round 3: 3HK and SmarTone retaliate — cut to HK$198")
print("Round 4+: Tit-for-Tat → Prices stabilize at HK$248")
print("\nResult: Temporary price war, then market equilibrium restored")
```

## 6. Summary: Game Theory in HK Markets
