## 1. Setup & Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 2. Supply & Demand Model — Hong Kong MTR Fare Analysis

### Case Study: MTR Fare Elasticity

MTR is a primary transportation mode in Hong Kong. Let's analyze fare sensitivity.

```python
# Supply & Demand Parameters (illustrative)
# Demand: P = a - b*Q (typical downward slope)
# Supply: P = c + d*Q (typical upward slope)

# Parameters
a = 12  # choke price
b = 0.02  # demand slope
c = 2  # minimum supply price
d = 0.01  # supply slope

# Calculate equilibrium
Q_eq = (a - c) / (b + d)
P_eq = a - b * Q_eq

print(f"Equilibrium: Q* = {Q_eq:.0f} million trips/day")
print(f"Equilibrium: P* = HK${P_eq:.2f}")
```

## 3. Elasticity Analysis

### Price Elasticity of Demand (PED)

$$PED = \frac{\% \Delta Q_d}{\% \Delta P}$$

If |PED| > 1: Elastic  
If |PED| = 1: Unitary  
If |PED| < 1: Inelastic

```python
# Calculate elasticity at equilibrium
# Using point elasticity method

def calculate_ped(P, Q, demand_slope):
    """Calculate price elasticity of demand"""
    return (demand_slope * P) / Q

PED = calculate_ped(P_eq, Q_eq, -b)
print(f"Price Elasticity of Demand at equilibrium: {PED:.3f}")

if abs(PED) > 1:
    print("→ Elastic demand")
elif abs(PED) == 1:
    print("→ Unitary elastic")
else:
    print("→ Inelastic demand")
```

## 4. HK Housing Market — Supply & Demand Analysis

Hong Kong housing market analysis using supply-demand framework.

```python
# HK Housing Market Case Study
# Illustrative data

hk_housing = pd.DataFrame({
    'Year': [2020, 2021, 2022, 2023, 2024],
    'Price_Index': [100, 105, 108, 112, 115],
    'Transactions': [70000, 65000, 60000, 58000, 55000]
})

print("Hong Kong Housing Market Data:")
print(hk_housing)

# Calculate price change
price_change = hk_housing['Price_Index'].pct_change() * 100
qty_change = hk_housing['Transactions'].pct_change() * 100

print(f"\nPrice change 2020-2024: {price_change.iloc[-1]:.1f}%")
print(f"Quantity change 2020-2024: {qty_change.iloc[-1]:.1f}%")
```

## 5. Visualization

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Supply & Demand
Q = np.linspace(0, 500, 100)
P_demand = a - b * Q
P_supply = c + d * Q

axes[0].plot(Q, P_demand, 'b-', label='Demand')
axes[0].plot(Q, P_supply, 'r-', label='Supply')
axes[0].axvline(x=Q_eq, color='g', linestyle='--', alpha=0.5)
axes[0].axhline(y=P_eq, color='g', linestyle='--', alpha=0.5)
axes[0].scatter([Q_eq], [P_eq], color='green', s=100, zorder=5)
axes[0].set_xlabel('Quantity')
axes[0].set_ylabel('Price (HK$)')
axes[0].set_title('Supply & Demand Equilibrium')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Plot 2: HK Housing Price Index
axes[1].plot(hk_housing['Year'], hk_housing['Price_Index'], 'o-')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Price Index')
axes[1].set_title('Hong Kong Housing Price Index')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('assets/hk_market_analysis.png', dpi=150)
plt.show()

print("Chart saved to assets/hk_market_analysis.png")
```
